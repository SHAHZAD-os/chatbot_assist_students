```python

from langchain_community.document_loaders import PyPDFLoader
from langchain_community.document_loaders import PyPDFDirectoryLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_community.vectorstores import FAISS

from langchain_community.embeddings import HuggingFaceBgeEmbeddings
from langchain.prompts import PromptTemplate

from langchain.chains import RetrievalQA

# Use PyPDFLoader to load single file
loader = PyPDFLoader(r"C:\Users\HB LAPTOP POINT\Desktop\New folder (2)\backend\BS Curriculm Computing Disciplines-2023.pdf")

documents = loader.load()

# Check if documents is empty
if not documents:
    print("No documents loaded. Check the file path and format.")
else:
    text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=10)
    final_documents = text_splitter.split_documents(documents)
    print(final_documents[0])
```

```python

from langchain_community.embeddings import HuggingFaceBgeEmbeddings

huggingface_embeddings=HuggingFaceBgeEmbeddings(
    model_name="sentence-transformers/paraphrase-MiniLM-L6-v2",      #sentence-transformers/all-MiniLM-l6-v2
    model_kwargs={'device':'cpu'},
    encode_kwargs={'normalize_embeddings':True}

)
```

```python
import  numpy as np
print(np.array(huggingface_embeddings.embed_query(final_documents[0].page_content)))
print(np.array(huggingface_embeddings.embed_query(final_documents[0].page_content)).shape)
```

```python
template = """
You are an AI assistant that provides answers based only on the context provided below.
Please carefully read the context and answer the question as specifically as possible.
Do not make assumptions beyond the information given.

Context:
{retrieved_info}

Question:
{question}

Answer:
"""





prompt = PromptTemplate.from_template(template)
```

```python
from langchain.chains import LLMChain
chain = LLMChain(prompt=prompt, llm=llm)
```

```python
# Set up the RetrievalQA chain
qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",  # Use appropriate chain type (stuff, map_reduce, etc.)
    retriever=vectorstore.as_retriever(),
    verbose=False
)

# Example question
question = "Enlist the domain core courses of comuter science?"

# Get relevant document from the vectorstore
relevant_documents = vectorstore.similarity_search(question, k=1)
document_content = relevant_documents[0].page_content if relevant_documents else "No relevant documents found."

# Format the final prompt
final_prompt = prompt.format(question=question, retrieved_info=document_content)

# Prepare the inputs for the chain (use 'query' if that's what the chain expects)
inputs = {
    "query": question,  # 'query' might be the expected key
    "retrieved_info": document_content  # Provide the context as 'retrieved_info'
}

# Get the answer from the chain
response = qa_chain.run(inputs)

# Print the response
print(response)
```
